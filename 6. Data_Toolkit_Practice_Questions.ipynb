3. Explain the following terms:

a. The difference in np.array, np.asarray and np.asanyarray.


1. np.array: Creates a new NumPy array from an input sequence. If the input is already a NumPy array, it returns a copy.
2. np.asarray: Converts an input sequence to a NumPy array. If the input is already a NumPy array, it returns a view (not a copy).
3. np.asanyarray: Similar to asarray, but allows more flexibility in data types and returns a view if possible.

b. The difference between Deep copy and shallow copy.

1. Shallow Copy: Creates a new object, but references the same objects as the original.
2. Deep Copy: Creates a new object and recursively copies all objects found in the original, creating entirely independent copies of all nested objects.

```python
original_array = np.array([1, 2, 3])
shallow_copy = original_array

print("Original array:", original_array)
shallow_copy[0] = 99
print("Original array:", original_array)

print("Shallow copy:", shallow_copy)


deep_copy = np.copy(original_array)
deep_copy[0] = 77

print("Original array:", original_array)
print("Deep copy:", deep_copy)
```

b) Extract all odd integers from array.

```python
# Generate a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

# Print the original array
print("Original Array:\n", random_array)

# Extract all even integers from the array
even_numbers = random_array[random_array % 2 == 0]

# Extract all odd integers from the array
odd_numbers = random_array[random_array % 2 != 0]

# Print even and odd arrays
print("\nEven Integers:", even_numbers)
print("\nOdd Integers:", odd_numbers)
```


14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
b) Label the x-axis as 'Date'.
c) Set the title of the plot as 'Temperature and Humidity Over Time'.


```python
# Generate sample time-series data
np.random.seed(0)
dates = pd.date_range('2022-01-01', periods=100)
temperature = np.random.normal(loc=25, scale=5, size=100)
humidity = np.random.normal(loc=50, scale=10, size=100)

# Create a DataFrame
data = {'Date': dates, 'Temperature': temperature, 'Humidity': humidity}
df = pd.DataFrame(data)

# Increase plot size
plt.figure(figsize=(20, 6))
# Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots(figsize=(20, 6))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show plot
plt.show()
```

15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
a) Plot a histogram of the data with 30 bins.
b) Overlay a line plot representing the normal distribution's probability density function (PDF).
c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability.
d) Set the title of the plot as 'Histogram with PDF Overlay'.


```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.5, color='b', label='Histogram')

# Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'r', linewidth=2, label='PDF')

# Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Show plot
plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data (replace with your own data if needed)
np.random.seed(42)
x_values = np.random.randn(100)  # Random x-values
y_values = np.random.randn(100)  # Random y-values

# Determine quadrant for each point
quadrant = np.zeros(100, dtype=int)  # Initialize with zeros
quadrant[(x_values > 0) & (y_values > 0)] = 1  # Quadrant I
quadrant[(x_values < 0) & (y_values > 0)] = 2  # Quadrant II
quadrant[(x_values < 0) & (y_values < 0)] = 3  # Quadrant III
quadrant[(x_values > 0) & (y_values < 0)] = 4  # Quadrant IV

# Create a DataFrame for plotting
data = {'x': x_values, 'y': y_values, 'Quadrant': quadrant}
df = pd.DataFrame(data)

# Set up the plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='Quadrant', data=df, palette='Set1', legend='full')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrant', loc='upper right')
plt.grid(True)
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate data for the sine wave function
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Add grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

# Show the plot
output_notebook()
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.


```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, HoverTool
import random

# Define categories and generate random data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = [random.randint(1, 10) for _ in range(len(categories))]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a Bokeh figure
p = figure(x_range=categories, height =450, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars to the plot
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='black', fill_color='values')

# Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Value", "@values")]
p.add_tools(hover)

# Set labels for axes and title
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Values"
p.xaxis.axis_label = "Categories"
p.title.align = "center"

# Show the plot
output_file("random_categorical_bar_chart.html")
show(p)
```
